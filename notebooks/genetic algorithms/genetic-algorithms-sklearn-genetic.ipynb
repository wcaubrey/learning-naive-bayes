```python
import pandas as pd

from sklearn_genetic import GAFeatureSelectionCV
from sklearn.tree import DecisionTreeClassifier
```

```python
"""
Load the dataset.
X will have the features.
Y will have the label we want to predict.
"""

df = pd.read_csv("../../data/heartdiseasetrain/heart.csv")
X = df.drop("target", axis=1)
y = df["target"]

num_features = X.shape[1]
feature_names = list(X.columns)
```

```python
CROSS_VALIDATION_SPLITTING = 5
SCORING_STRATEGY = "accuracy"
POPULATION_SIZE = 30
NUM_GENERATIONS = 50
MUTATION_RATE = 0.05
NUM_JOBS = 1
RANDOM_STATE = 42
```

```python
selector = GAFeatureSelectionCV(
    estimator=DecisionTreeClassifier(random_state=RANDOM_STATE),
    cv=CROSS_VALIDATION_SPLITTING,
    scoring=SCORING_STRATEGY,
    population_size=POPULATION_SIZE,
    generations=NUM_GENERATIONS,
    mutation_probability=MUTATION_RATE,
)
selector.fit(X, y)
```

```python
# Best feature names (if X is a DataFrame with column names)
print("Best feature subset:", selector.get_feature_names_out(X.columns))

print(f"Best Accuracy: {max(selector.history['fitness_max']):.4f}")
```
